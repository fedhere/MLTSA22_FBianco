```python
import numpy as np
from numpy import pi, cos, linspace, arange
from numpy.fft import rfft, irfft
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
```

```python
import scipy as sp

from scipy import stats
```

```python
sp.stats.norm.rvs?
```

```python
np.random.seed = 123

frames = 1000
mygauss = []
for i in range(frames):
  mygauss.append(sp.stats.norm.rvs(loc = 100, scale = 10))
```

```python

fig = plt.figure()
fig.set_dpi(100)
ax = fig.add_subplot(1,1,1)

def animate(t):
    plt.xlim([70., 130.0])
    p, = plt.plot(mygauss[t], 0, 'ko', alpha=0.2)
anim = animation.FuncAnimation(fig, animate, frames=frames, interval=100, repeat=False)
HTML(anim.to_jshtml())



  
```

```python

fig = plt.figure()
fig.set_dpi(100)
ax = fig.add_subplot(1,1,1)

def animatehist(t):
    plt.xlim([70., 130.0])
    plt.ylim([-1., 10.0])
    plt.hist(mygauss[:t], bins=np.arange(-80,120,3), color='k', alpha=0.3)
animhist = animation.FuncAnimation(fig, animatehist, frames=frames//3, interval=100, repeat=False)
HTML(animhist.to_jshtml())
```
