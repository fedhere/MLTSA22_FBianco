```python
import pylab as pl
import numpy as np
```

```python
pl.figure(figsize=(20,5))
N = 200
np.random.seed(100)
y = np.random.randn(N) * N
t = np.linspace(0, N, N, endpoint=False)
pl.plot(t, y, lw=2)
pl.xlabel("time")
pl.ylabel("y");
```

```python
pl.figure(figsize=(20,5))
N = 200
np.random.seed(100)
y = np.random.randn(N)  * np.arange(N)
t = np.linspace(0, N, N, endpoint=False)
pl.plot(t, y, lw=2)
pl.xlabel("time")
pl.ylabel("y");
```
